```python
import pandas            as pd
import numpy             as np
import cv2
import os
import matplotlib.pyplot as plt
import glob
from collections import Counter
```

```python
def sumMasks(mask_list):

    """
    This function sums a list of given masks.
    Parameters
    ----------
    mask_list : {list of numpy.ndarray}
        A list of masks (numpy.ndarray) that needs to be summed.
    Returns
    -------
    summed_mask_bw: {numpy.ndarray}
        The summed mask, ranging from [0, 1].
    """

    try:
        # import pdb; pdb.set_trace()
        summed_mask = np.zeros(mask_list[0].shape)

        for arr in mask_list:
            summed_mask = np.add(summed_mask, arr)

        # Binarise (there might be some overlap, resulting in pixels with
        # values of 510, 765, etc...)
        _, summed_mask_bw = cv2.threshold(
            src=summed_mask, thresh=1, maxval=255, type=cv2.THRESH_BINARY
        )

    except Exception as e:
        print((f"Unable to get findMultiTumour!\n{e}"))

    return summed_mask_bw
```

```python
# multi masks

#train images
multi_masks_paths_train = {}
for p, count in zip(temp_train.index,temp_train.values):
    # print(p, count)
    paths = []
    for i in range(1, count+1):
        path = "{}_MASK_{}.png".format(p, i)
        paths.append(path)
        # print(path)
    if count>1:
        multi_masks_paths_train[p] = paths
        
# test images
multi_masks_paths_test = {}
for p, count in zip(temp_test.index,temp_test.values):
 
    print(p, count)
    paths = []
    for i in range(1, count+1):
        path = "{}_MASK_{}.png".format(p, i)
        paths.append(path)
        print(path)
    if count>1:
        multi_masks_paths_test[p] = paths
    
```

```python
#Training images
input_path = "/datasets/mamografia/CBIS-DDSM_organized/images/preprocessed/train/mask"
output_path = "/datasets/mamografia/CBIS-DDSM_organized/images/preprocessed/train/merged_masks/"
counter = 0
for k, v in multi_masks_paths_train.items():
    print("image {} ...\n".format(counter))
    # Get image as arrays.
    mask_list = [cv2.imread(os.path.join(input_path, "_".join(["Mass-Training",mask_path])), cv2.IMREAD_GRAYSCALE) for mask_path in v]
    print(type(mask_list[0]))
    print([os.path.join(input_path, mask_path) for mask_path in v])
    # import pdb;pdb.set_trace()
    
    # Sum masks
    summed_mask = sumMasks(mask_list=mask_list)

    # Save summed mask

    save_path = os.path.join(
                output_path, ("_".join(["Mass-Training", k, "MASK_MERGED.png"]))
            )
    cv2.imwrite(save_path, summed_mask)
    counter += 1
```

## Training

```python
import glob

input_path = "/datasets/mamografia/CBIS-DDSM_organized/images/preprocessed/train/mask/"
files = glob.glob('{}/*.png'.format(input_path))
```

```python
len(files)
```

```python
files[:3]
```

```python
cut_files = [f[:-5] for f in files]
```

```python
cut_files[:3]
```

```python

contador = 0

freq_files = Counter(cut_files)
# for k,v in Counter(cut_files).items():
#     if v>1:
#         contador+=1
print(len(freq_files))
```

```python
dict_paths = {}

for i in cut_files:
    dict_paths[i] = []
    
for i in files:
    cut_name = i[:-5]  # remove number.png
    # print(cut_name)
    dict_paths[cut_name].append(i)

```

```python
len(dict_paths)
```

```python
from shutil import copyfile
output_path = "/datasets/mamografia/CBIS-DDSM_organized/images/preprocessed/train/merged_masks/"

contador=0
for k, v in dict_paths.items():
    if len(v)>1:
        
        mask_list = [cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE) for mask_path in v]
        # Sum masks
        summed_mask = sumMasks(mask_list=mask_list)
        image_name_cut = k.split('/')[-1]
        save_path = os.path.join(
                output_path, ("".join([image_name_cut,"MERGED.png"]))
            )
        cv2.imwrite(save_path, summed_mask)
        
        # print(save_path)
    else:
        #copy image
        # 
        src = v[0]
        
        filename= v[0].split('/')[-1]
        dst = "".join([output_path, filename])
        # print(dst)
        copyfile(src, dst)
    contador+=1
    
print(contador)
        
```
