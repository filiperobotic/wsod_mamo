```python
#imports
import pandas as pd
import glob
```

```python
df_train = pd.read_csv("/datasets/mamografia/CBIS-DDSM_organized/csv-description/mass_case_description_train_set.csv")
df_test = pd.read_csv("/datasets/mamografia/CBIS-DDSM_organized/csv-description/mass_case_description_test_set.csv")
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
new_cols = [col.replace(" ", "_") for col in df_train.columns]
df_train.columns = new_cols

new_cols = [col.replace(" ", "_") for col in df_test.columns]
df_test.columns = new_cols


df_train.head()
```

```python
df_train["identifier"] = df_train.apply(lambda x: "_".join([x["patient_id"], x["left_or_right_breast"], x["image_view"]]), axis=1)
df_test["identifier"] = df_test.apply(lambda x: "_".join([x["patient_id"], x["left_or_right_breast"], x["image_view"]]), axis=1)
df_train.head()
```

```python
# Keep only the images with mass abormality type

#mass_train = df_train.loc[df_train["abnormality_type"] == "mass"]["identifier"].value_counts()
df_mass_train = df_train.loc[df_train["abnormality_type"] == "mass"]#["identifier"]#.value_counts()
df_mass_test = df_test.loc[df_train["abnormality_type"] == "mass"]#["identifier"].value_counts()
```

```python
df_mass_train.head()
```

```python
# Set the image path for png files

df_mass_train["image_full_png_path"] = df_mass_train.apply(lambda x: "Mass-Training_" + x["identifier"] + "_FULL.png", axis=1)
df_mass_test["image_full_png_path"] = df_mass_test.apply(lambda x: "Mass-Test_" + x["identifier"] + "_FULL.png", axis=1)
```

```python
df_mass_train.head()
```

```python
temp_train = df_mass_train["identifier"].value_counts()
temp_test = df_mass_test["identifier"].value_counts()
temp_train
```

```python
temp_test
```

```python
# Set the png mask image path
train_masks_path = "/datasets/mamografia/CBIS-DDSM_organized/images/preprocessed_gmic/train/merged_masks/"
test_masks_path = "/datasets/mamografia/CBIS-DDSM_organized/images/preprocessed_gmic/test/merged_masks/"
df_mass_train["image_mask_png_path"] = df_mass_train.apply(lambda x: glob.glob('{}/*{}*.png'.format(train_masks_path,x["identifier"]))[0].split("/")[-1], axis=1)
df_mass_test["image_mask_png_path"] = df_mass_test.apply(lambda x: glob.glob('{}/*{}*.png'.format(test_masks_path,x["identifier"]))[0].split("/")[-1], axis=1)
```

```python
# Set the png mask image path

# df_mass_train["image_mask_png_path"] = df_mass_train.apply(lambda x: "Mass-Training_" + x["identifier"] + "_MASK_MERGED.png" if temp_train[x["identifier"]]>1 else "Mass-Training_" + x["identifier"] + "_MASK_1.png", axis=1)
# df_mass_test["image_mask_png_path"] = df_mass_test.apply(lambda x: "Mass-Training_" + x["identifier"] + "_MASK_MERGED.png" if temp_test[x["identifier"]]>1 else "Mass-Training_" + x["identifier"] + "_MASK_1.png", axis=1)
```

```python
# Rearrange the coluns to the identifier be the first column
df_mass_train = df_mass_train[ ['identifier'] + [ col for col in df_mass_train.columns if col != 'identifier' ] ]
df_mass_test = df_mass_test[ ['identifier'] + [ col for col in df_mass_test.columns if col != 'identifier' ] ]
```

```python
# Remove lines with duplicades identifier. Once we merged the maks, we must remove the other rows

df_mass_train = df_mass_train.drop_duplicates(subset=['identifier'], keep='first')
df_mass_test = df_mass_test.drop_duplicates(subset=['identifier'], keep='first')
df_mass_train
```

```python
df_mass_test
```

```python
# Add binary pathology column 
df_mass_train['binary pathology'] = df_mass_train.apply(lambda x: 1 if x["pathology"]=="MALIGNANT" else 0, axis=1)
df_mass_test['binary pathology'] = df_mass_test.apply(lambda x: 1 if x["pathology"]=="MALIGNANT" else 0, axis=1)
```
